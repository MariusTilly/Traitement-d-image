```python
import PIL
```

```python
from PIL import Image
toto = Image.open("photo/thermographie-pont-thermique.jpg")
```

```python
print(toto.size, toto.format, toto.mode)
```

```python
toto
```

```python
for i in range (0,400) :
    for j in range (0,300):
        r=toto.getpixel((i,j))[1]
        toto.putpixel((i,j),(r,r,0))
```

```python
toto
```

```python
toto.save("resultat.png", format="png")
```
